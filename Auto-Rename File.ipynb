##### CODE DƯỚI ĐÂY SỬ DỤNG CHO CÁC FILE DẠNG GẦN CHUẨN
    VD: SSB_Baocaotaichinh_Q1_2020_Congtyme_Soatxet.pdf
    --> SSB_BCTC_2020_Q1_MXS.pdf

```python
# Hàm thay đổi về tên viết tắt
def convert_to_abbreviation(part):
    abbreviations = {
        'Baocaoquantri': 'BCQT',
        'Baocaotaichinh': 'BCTC',
        'Bancaobach': 'BCB',
        'Tomtat': 'TT',
        'Kiemtoan': 'KT',
        'Soatxet': 'XS',
        'NiemYet': 'NY',
        'Congtyme': 'M',
        'Hopnhat': 'HN',
        'PhatHanhThem': 'PHT',
        'DieuChinh': 'DC'
    }
    return abbreviations.get(part, part)



# Hàm kiểm tra xem 2 chuỗi cuối có cùng nằm trong 1 danh sách các từ viết tắt ko? 
# Nếu có thì hợp nhất 2 từ đó lại thành quy chuẩn chung 
def check_the_last_2string(parts_list):
    if len(parts_list) <2:
        return 'False'
    
    list_string_to_check = ['Soatxet', 'Congtyme', 'Hopnhat', 'Kiemtoan']
    end_string = parts_list[-1]
    next_to_end = parts_list[-2]
    
    if (end_string and next_to_end) in list_string_to_check:
        for string in list_string_to_check:
            if string == end_string:
                end_string = convert_to_abbreviation(string)
            if string == next_to_end:
                next_to_end = convert_to_abbreviation(string)
        result = end_string + next_to_end      
        return result
    
    else:
        return 'False'
    

   
#############################################################################################################################



# Tạo list gồm tên 3 cái thư mục
list_folder_name = ['Báo cáo tài chính', 'Báo cáo quản trị', 'Bản cáo bạch và báo cáo thường niên']
banking_name = 'STB'


# Thư mục(Folder) chứa các file(tệp) muốn đổi tên
folder_path = f'D:\Thực tập\Công việc\My Work\pdf_files'


# Lấy danh sách tất cả các file trong thư mục
file_list = os.listdir(folder_path)


# Lặp qua từng tệp và đổi tên
for index, file_name in enumerate(file_list):
    # Lấy phần tên của tệp ra và chia ra thành các phần chính của 
    name_only, extension_ = os.path.splitext(file_name)
    parts_list = name_only.split('_')
    
    
    # Kiểm tra xem 2 chuỗi cuối trong list có riêng lẻ ko và cần phải combine hay ko
    check = check_the_last_2string(parts_list)
    new_file_name = parts_list[0]
    # if new_file_name == 'STB':
    #     new_file_name = 'STB'
    
    if check == 'False':
        # Tạo tên mới cho tệp
        for index, part in enumerate(parts_list[1:]): 
            if index == 1:
                if len(part) == 2:
                    new_file_name = new_file_name + '_' + parts_list[3] + '_' + part
                    continue
            if index == 2:
                continue
            str_ = convert_to_abbreviation(part)
            new_file_name = new_file_name + '_' + str_
        new_file_name += extension_
    else:
        for index, part in enumerate(parts_list[1:-2]): 
            if index == 1:
                if len(part) == 2:
                    new_file_name = new_file_name + '_' + parts_list[3] + '_' + part
                    continue
            if index == 2:
                continue
            str_ = convert_to_abbreviation(part)    
            new_file_name = new_file_name + '_' + str_
        new_file_name = new_file_name + '_' + check + extension_     
            
            
    # Tạo đường dẫn đến tệp cũ và mới
    old_path = os.path.join(folder_path, file_name)
    new_path = os.path.join(folder_path, new_file_name)
    
            
    # Đổi tên tệp
    os.renames(old_path, new_path)
```
